```python
import csv
import sys
# set the csv file limt to be the maximum
csv.field_size_limit(sys.maxsize)

# Open the CSV file
id_dict = {}
input_file = "pre-citation-04-filtered-data.csv"
output_file = "04output.csv"
with open(input_file, 'r') as file:
    # Create a CSV reader object
    reader = csv.reader(file)

    # Iterate over each row in the CSV file
    index = 0
    for row in reader:
        # Process the row data
        id_dict[row[0]] = index
        index += 1

file = open(output_file, 'w')
isHeader=True
with open(input_file, encoding="utf8") as csvfile:
  
        csvreader = csv.reader(csvfile, delimiter=',')
        # Iterate over each row in the CSV file
        for row in csvreader:
            if isHeader: # add the header
                writer = csv.writer(file)
                writer.writerow(row)
                isHeader = False
            else:
                # Process the row data
                arr_refs = row[22].strip().split()
                if len(arr_refs) != 0:
                    my_row = row
                    my_row[0] = id_dict[row[0]] # convert the pmid
                    string_ref = ""
                    
                    # convert the references
                    for ref in arr_refs:
                        id = str(ref)
                        if id in id_dict:
                            if(my_row[0] != id_dict[id]):
                                string_ref += str(id_dict[id]) + " "
                            else:
                                print(f"for pmid = {my_row[0]}: new ref ={id_dict[id]}")
                    
                    my_row[22] = string_ref
                    writer = csv.writer(file)
                    writer.writerow(my_row)
                    
                    
file.close()
```
